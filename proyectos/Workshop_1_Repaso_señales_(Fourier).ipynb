```python
import numpy as np
import matplotlib.pyplot as plt

def fourier_coefficients(f, t, T, k_max):
    kp = k_max//2
    C_k = []
    for k in range(-kp,kp):
        omega_k = 2 * np.pi * k / T
        integrand = f * np.exp(-1j * omega_k * t)
        C_k.append((1/T) * np.trapz(integrand, t))
    return np.array(C_k)

T = 1
t = np.linspace(0, T, 500, endpoint=False)
frecuencia = 5
señal_pulsante = (np.sign(np.sin(2 * np.pi * frecuencia * t))) + 1


k_max = 200
C_k = fourier_coefficients(señal_pulsante, t, T, k_max)

plt.figure()

plt.subplot(2, 1, 1)
plt.plot(t, señal_pulsante)
plt.title("Señal Pulsante (Onda Cuadrada)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(range(k_max), np.abs(C_k), 'b', markerfmt=" ", basefmt="-b")
plt.title("Coeficientes de Fourier")
plt.xlabel("Armónico k")
plt.ylabel("|C_k| (Magnitud)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 1
fs = 1000
t = np.linspace(0, 5 * T, 5 * T * fs)

def shark_fin_wave_exp(t, T):
    tau = t % T
    if tau < 0.3 * T:
        return 1 - np.exp(-15 * tau / T)
    else:
        return np.exp(-6 * (tau - 0.3 * T) / T)

shark_wave_exp = np.array([shark_fin_wave_exp(ti, T) for ti in t])

def fourier_coefficients(f, t, T, k_max):
    kp = k_max // 2
    C_k = []
    for k in range(-kp, kp):
        omega_k = 2 * np.pi * k / T
        integrand = f * np.exp(-1j * omega_k * t)
        C_k.append((1 / T) * np.trapz(integrand, t))
    return np.array(C_k)

k_max = 80
C_k = fourier_coefficients(shark_wave_exp, t, T, k_max)

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, shark_wave_exp)
plt.title("Señal Periódica con Forma de Aleta de Tiburón (Exponencial)")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(np.arange(-k_max//2, k_max//2), np.abs(C_k), basefmt=" ", use_line_collection=True)
plt.title("Coeficientes de Fourier")
plt.xlabel("k")
plt.ylabel("|C_k|")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
T = 1  # Periodo
fs = 1000  # Frecuencia de muestreo
t = np.linspace(0, 5 * T, 5 * T * fs)  # Tiempo

# Función de onda triangular sin usar numpy.tri
def triangular_wave(t, T):
    tau = t % T
    if tau < 0.5 * T:  # Subida
        return 2 * tau / T
    else:  # Bajada
        return 2 - 2 * tau / T

# Generar la señal triangular
triangular_wave_signal = np.array([triangular_wave(ti, T) for ti in t])

# Función para calcular los coeficientes de Fourier
def fourier_coefficients(f, t, T, k_max):
    kp = k_max // 2
    C_k = []
    for k in range(-kp, kp):
        omega_k = 2 * np.pi * k / T
        integrand = f * np.exp(-1j * omega_k * t)
        C_k.append((1 / T) * np.trapz(integrand, t))
    return np.array(C_k)

# Calcular coeficientes de Fourier
k_max = 80
C_k_triangular = fourier_coefficients(triangular_wave_signal, t, T, k_max)

# Graficar la señal triangular y los coeficientes de Fourier
plt.figure(figsize=(12, 6))

# Gráfico de la señal triangular
plt.subplot(2, 1, 1)
plt.plot(t, triangular_wave_signal)
plt.title("Señal Triangular")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)

# Gráfico de los coeficientes de Fourier
plt.subplot(2, 1, 2)
plt.stem(np.arange(-k_max//2, k_max//2), np.abs(C_k_triangular), basefmt=" ", use_line_collection=True)
plt.title("Coeficientes de Fourier (Señal Triangular)")
plt.xlabel("k")
plt.ylabel("|C_k|")
plt.grid(True)

plt.tight_layout()
plt.show()
```
