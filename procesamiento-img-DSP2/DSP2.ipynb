```python
def plot_hist(intensity,freq):
    fig, ax = plt.subplots()
    fig.set_size_inches((15,7))
    ax.bar(intensity,freq)
    plt.show()
```

```python
def image_histogram(x):
  intensity = np.linspace(0,255, 256)
  freq = np.zeros(256)
  for i in range (x.shape[0]):
    for j in range (x.shape[1]):
      freq[x[i,j]] += 1
  plot_hist(intensity,freq)
  return intensity, freq
```

```python
image_histogram(xr)
```
